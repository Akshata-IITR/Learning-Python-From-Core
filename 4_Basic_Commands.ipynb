```python
# Enumeratre in python

marks =[1,12,13,17,41,98,20,68]

for index,mark in enumerate(marks,start=0): 
    # inplace of zero we can put desired index from which we want to start
    print(mark)
    if index==5:
        print("Awesome")
    index +=1
```
