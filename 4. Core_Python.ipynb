```python
# "*" will treat variable num as a tuple

def average(*num):
    sum=0
    for i in num:
        sum=sum+i
    print("Average is: ", sum/len(num))

average(1, 2, 3)
```

```python
# List methods

l=[1,3,2,5]

# 1. Arranging elements in  lis into ascending order
# Changes original list
l.sort()
print(l)

# 2. Arranging in descending order
# Changes original list

l.sort(reverse=True)
print(l)

# 3. Reverse elements on list
#Changes original list
l.reverse()
print(l)

# 4. Finding index of a element in a list
# will give index of first occured element if there are multiple elements with same value

print(l.index(5))

# 5. Counting occurance of a element

print(l.count(2))

# 6. Copy list l and make changes is m 
# original list will not be changed

m=l.copy()
m.reverse()
print(l)
print(m)


# 7. Replacing element on a given index of a list
# Changes original list

l.insert(1,"x")
print(l)

# Merge two lists
# order of addition matters

k=l+m
print(k)

# extending original list l adding all elemnts of m 
l.extend(m)
print(l)
```

```python
# String formatting 
# in A it will take what we have gave in order 
# in B it will take first input as second and vice versa 
A="Hey my name is {} and i am from {}"
B="HEy my name is {1} and i am from {0}"

Country="India"
Name="Akshata"

A_1=A.format(Name, Country)
A_2=A.format(Country, Name)
B_1=B.format(Name, Country)
B_2=B.format(Country, Name)
print("1. ",A_1)
print("2. ", A_2)
print("3. ",B_1)
print("4. ",B_2)


# OR use f-string

print(f"Hey my name is {Name} and i am from {Country}")

# if we want upto 'p' decimal point then write
# {Name:.pf} where 'p' is any number

txt="for only {price:.4f} dollars "
print(txt.format(price=49.00999)) # upto 4 decimal

# OR

price=52.07890
text=f"for only {price:.3f} dollars "
print(text)

# If we want a curly bracket inside f string
# then make a double curly vracket it will be printed
# as a single curly bracket without introducing variable vallue 
print(f"for only {{price:.3f}} dollars ")
```

```python
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(4))
```
